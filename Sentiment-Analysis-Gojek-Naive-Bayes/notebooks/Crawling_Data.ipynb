#Import library


```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
#scrape all available reviews
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
'com.gojek.app',
lang = 'id',
country ='id',
sort=Sort.NEWEST,
count=10000,
filter_score_with=None

)
```

#Show Data yang sudah di crawling

```python
df_busu = pd.DataFrame(np.array(result),columns=['review']) #nama file bebas
df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))
df_busu.tail()
```
