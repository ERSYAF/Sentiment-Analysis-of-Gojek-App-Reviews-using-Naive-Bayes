#Memasukkan Dataset

```python
df = pd.read_csv('/content/label_gojek.csv')
```

#Praproses Data

```python
#konversi ke lowercase
df['content'] = df['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df.head()
```

#Menghapus angka,simbol dan emoji

```python
#Menghilangkan angka
df['content'] = df['content'].str.replace('[0-9]+','')

#Menghilangkan punctuations = tandabaca ex:@,_.#,etc dan emoji
df['content'] = df['content'].str.replace('[^\w\s]+','',)
df.head(30)
```

#Removing Stopwords

```python
#stopwords menggunakan sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopwords_id = factory.get_stop_words()

def remove_stopwords(text):
  """custom function to remove the stopwords"""
  return " ".join([word for word in str(text).split() if word not in stopwords_id])

df["content"] = df["content"].apply(lambda text:remove_stopwords(text))
df.head()
```

#Stemming

```python
from Sastrawi.Stemmer.Stemmer import Stemmer
#Stemming Bahasa Indonesia
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
Stemmer_id = factory.create_stemmer()

def stem_words(text):
  return " ".join([Stemmer_id.stem(word) for word in text.split()])

df["content"] = df["content"].apply(lambda text: stem_words(text))
df.head()
```

#Tokenize

```python
#Tokenization bertujuan mengubah kata ke dalam unit terkecil bermakna
def tokenize_text(text):
  tokenized_text = text.split()
  return tokenized_text
df['content'] = df['content'].apply(lambda x:tokenize_text(x))
df.head()
```

#Lemmatize

```python
#Lemmatization
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
def lemmatize_text(token_list):
    return " ".join([lemmatizer.lemmatize(token) for token in token_list if not token in set()])

lemmatizer = nltk.stem.WordNetLemmatizer()
df['content'] = df['content'].apply(lambda x: lemmatize_text(x))
df.head()
```
